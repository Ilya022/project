4. Нахождение дат:

Отбор по шаблону даты: dd-mm-yyyy или dd/mm/yyyy

```python

import re
pattern = r'\b\d{1,2}[-/]\d{1,2}[-/]\d{4}\b'
dates = "11-11-2023, 22/12/2222, 16 февраля 2000 года"
matches = re.findall(pattern, dates)
print(matches)
```

5. Нахождение цен:

Цена отбирается по слову "рубля" с возможностью склонения, а так же по цифрам.

```python
import re
pattern = r'\b\d+(?:\.\d{1,2})?\s?руб(?:ля|лей)?\b'
money = "100 рублей, 200 рублей, 300 рублей, 400 долларов, 500 евро"
matches = re.findall(pattern, money)
print(matches)
```

6. Нахождение текста в HTML тегах.

Текст выводится с помощью ленивого поиска текста между символами < и >.


```python
import re
html_text = '<p> Mercedes-Benz ([mɛʁˈtseːdəs ˈbɛnts] или [-dɛs-]; рус. Мерсе́дес-Бенц) — торговая марка и одноимённая компания — производитель легковых автомобилей премиального класса, грузовых автомобилей, автобусов и других транспортных средств, входящая в состав немецкого концерна «Mercedes-Benz Group». Является одним из самых узнаваемых автомобильных брендов во всём мире. Штаб-квартира Mercedes-Benz находится в Штутгарте, Баден-Вюртемберг, Германия. </p>'
pattern = re.compile(r'<[^>]*>(.*?)<[^>]*>')
matches = pattern.findall(html_text)
print(matches)
```

7. Нахождение IP адресов V4:

Отбор ведется с помощью шаблона: "ccc.ссс.с.с"

```python
import re
pattern = r'\b(?:\d{1,3}\.){3}\d{1,3}\b'
ip = "111.111.1.1, 222.222.0.233, 100.13.31.5, 12312312312, одиндватри.четырепятьшесть.7.8, 1.111.11111.3"
matches = re.findall(pattern, ip)
print(matches)
```

8. Определение протокола в URL:

Отбор протокола ведется до "://", включающий буквы.

```python

import re
pattern = r'^([a-zA-Z]+)://'
url = "https://www.123.com"
matches = re.findall(pattern, url)
print(matches)
```

9. Нахождение всех Московских городских номеров

Нахождение производится при условии, что номер начинается с "495".

```python

import re
pattern = r'\(495\)\s\d{3}-\d{4}'
text = "(495) 222-2222, (495) 111-1111, 123123, (925) 124-1244, 98776123"
matches = re.findall(pattern, text)
print(matches)
```
